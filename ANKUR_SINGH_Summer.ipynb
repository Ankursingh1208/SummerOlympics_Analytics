```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv('summer.csv')
df
```

```python
df.isnull().sum()
```

# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
len(df["City"].unique())

```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python

df[df.Medal == 'Gold'].groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5),color=['black', 'red', 'green', 'blue', 'cyan'])

```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5),color=['black', 'red', 'green', 'blue', 'cyan'])
```

### 4. Which player has won most number of medals? (Top 5)

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (10,5),color=['black', 'red', 'green', 'blue', 'cyan'])
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df[df.Medal == 'Gold'].groupby('Athlete').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Athlete', y = 'freq', kind = 'bar', figsize = (10,5),color=['black', 'red', 'green', 'blue', 'cyan'])
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df[(df.Medal == 'Gold') & (df.Country=='IND')].Year.head(1)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby('Event').count()['Athlete'].sort_values(ascending = False).head().plot(x = 'Event', y = 'freq', kind = 'bar', figsize = (10,5),color=['black', 'red', 'green', 'blue', 'cyan'])
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df[(df.Medal == 'Gold') & (df.Gender=='Women')].groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5),color=['black', 'red', 'green', 'blue', 'cyan'])
```
